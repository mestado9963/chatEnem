```python
import sys

# Adiciona o diretório raiz ao Python path
sys.path.append('../')

from chat_llm import ChatLLM
from chat_enem_tutor import ChatLLMTutor
from langchain.chat_models import init_chat_model



```

```python
from huggingface_hub import InferenceClient
import os

tutors_chats = []
persons_chats = []

for llm in llms:
    print(f"Model: {llm.name}, Tag: {llm.tag}, Environment Key: {llm.environmentKey}")

    ## Inicializa o chat
    chat_tutor = ChatLLMTutor()
    chat_person = ChatLLM()

    # Configurar modelo
    chat_tutor.model = llm
    chat_person.model = llm

    # Criar cliente InferenceClient
    client_tutor = InferenceClient(
        model=chat_tutor.model.name,
        token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
    )

    client_person = InferenceClient(
        model=chat_person.model.name,
        token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
    )

    # Atribuir clientes às instâncias das classes
    chat_tutor.llm = client_tutor
    chat_tutor.temperature = 0.2

    chat_person.llm = client_person
    chat_person.temperature = 0.2

    # Definir a memória
    chat_tutor.memory = []
    chat_person.memory = []

    tutors_chats.append(chat_tutor)
    persons_chats.append(chat_person)
```

```python
from datasets import load_dataset, Features, Value

# Caminho para o diretorio
json_dir_path = "../3_datasets/enem/"

features = Features({
    "ano": Value("int64"),
    "enunciado": Value("string"),
    "label": Value("int64"),
    "cor_prova": Value("string"),
    "pos_item": Value("string"),
    "descricao_area": Value("string"),
    "area": Value("string"),
    "cod_prova": Value("int64"),
    "CO_POSICAO": Value("float64"),
    "SG_AREA": Value("string"),
    "CO_ITEM": Value("float64"),
    "TX_GABARITO": Value("string"),
    "CO_HABILIDADE": Value("float64"),
    "IN_ITEM_ABAN": Value("float64"),
    "TX_MOTIVO_ABAN": Value("string"),
    "NU_PARAM_A": Value("float64"),
    "NU_PARAM_B": Value("float64"),
    "NU_PARAM_C": Value("float64"),
    "TX_COR": Value("string"),
    "CO_PROVA": Value("float64"),
    "TP_LINGUA": Value("int64"),
    "IN_ITEM_ADAPTADO": Value("float64")
})

anos = ["2024"]

enem_datasets = {}

for ano in anos:
    path_ano = f"{json_dir_path}/{ano}"
    enem_datasets[ano] = load_dataset(
        "json",
        data_files={
            "train": f"{path_ano}/train/**/*.json",
            "test": f"{path_ano}/test/**/*.json",
            "full": f"{path_ano}/**/*.json"
        },
        features=features
    )

# Exibir algumas amostras do dataset
print(enem_datasets)
```

```python
# Carregar o arquivo ../persons_extraction/persons_se.json.
import json
import os


persons = '../4_persons_extraction/matriz_referencia/persons.json'

resp_person = []

with open(persons, 'r', encoding='utf-8') as f:
    data = json.load(f)

    #Para os primeiros 3 itens do dataset, imprimir o nome e a descrição.
    for index, person in enumerate(data):

        habilities = ""

        for skill in person["skills"]:
            # concatenar habilidades em uma string separada por quebras de linha.
            habilities_list = "\n\t\t\t- ".join(skill["HABILIDADES"])
            habilities += f"""
                        Para questões do ENEM da área de conhecimento "{skill["AREA_CONHECIMENTO"]}" \n
                        Você possui a competência "{skill["COMPETENCIA"]}"  \n
                        que acompanha das seguintes habilidades: \n
                        - {habilities_list} \n
                    """
            #print(habilities)
        for chat_person in persons_chats:
            #print(chat_person.model.tag)
            #Definir a persona
            chat_person.skills = habilities

            questions_serializable = []
            responses_questions = []

            for question in questions:

                chat_person.questions = question["enunciado"]

                state = chat_person.get_response({
                    "memory": []
                    })

                # print(state["answer"])

                question_serializable = {
                    "enunciado": question["enunciado"],
                    "ano": question["ano"],
                    "descricao_area": question["descricao_area"],
                    "CO_POSICAO": question["CO_POSICAO"]
                }
                questions_serializable.append(question_serializable)
                
                resp_question = {
                    "ano": question["ano"],
                    "tag_model": chat_person.model.tag,
                    "CO_POSICAO": question["CO_POSICAO"],
                    "descricao_area": question["descricao_area"],
                    "resposta": state["answer"]
                }

                responses_questions.append(resp_question)

            resp = {
                'id_person' : person["id"],
                'tag_model_person': chat_person.model.tag,
                'questions': questions_serializable,
                'resp': responses_questions
            }

            #print(resp)
            # # Criar um campo "resp" com os valores retornados por state["answer"].
            resp_person.append(resp)
            #print(resp_person)

    # Crie um arquivo .json que armazene os dados da variável item.
    # Salvar o dataset de participantes em um arquivo JSON

    # # Criar diretório se não existir
    dataset_dir = './'
    os.makedirs(dataset_dir, exist_ok=True)

    # Salvar o dataset em JSON
    output_path = os.path.join(dataset_dir, f'persons_resp.json')
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(resp_person, f, ensure_ascii=False, indent=2)
```

```python
#Gerando os feedbacks de cada resposta.

# Carregar o arquivo ../persons_extraction/persons_se.json.
import json
import os


resp_person = '../5_experiments/persons_resp.json'

resp_tutor = []

with open(resp_person, 'r', encoding='utf-8') as f:
    data = json.load(f)

    #Para os primeiros 3 itens do dataset, imprimir o nome e a descrição.
    for index, person in enumerate(data):
        
        for chat_tutor in tutors_chats:
            questions_serializable = []
            feedbacks_questions = []
            
            for index, question in enumerate(questions):
                #print("resposta: ", person["resp"][index]["resposta"])
                chat_tutor.persona_resp = person["resp"][index]["resposta"]
                chat_tutor.questions = question["enunciado"]
                
                state = chat_tutor.get_response({
                    "memory": []
                    })
                
                feedback_question = state["answer"]
                #print(state["answer"])

                question_serializable = {
                    "enunciado": question["enunciado"],
                    "ano": question["ano"],
                    "descricao_area": question["descricao_area"],
                    "SG_AREA": question["SG_AREA"],
                    "CO_POSICAO": question["CO_POSICAO"],
                    "TP_LINGUA": question["TP_LINGUA"],
                    "TX_GABARITO": question["TX_GABARITO"],
                    "CO_HABILIDADE": question["CO_HABILIDADE"],
                    "NU_PARAM_A": question["NU_PARAM_A"],
                    "NU_PARAM_B": question["NU_PARAM_B"],
                    "NU_PARAM_C": question["NU_PARAM_C"]
                }
                questions_serializable.append(question_serializable)

                # print(len(feedbacks_person))
                #print(feedback_question)

                if feedback_question.strip() != "":
                    feedback_split = feedback_question.split('[##]')
                    numero_questao = feedback_split[0].strip()
                    resposta = feedback_split[1].strip()
                    feedback_tutor = feedback_split[2].strip()

                feedbacks_questions.append({
                    "numero_questao": numero_questao,
                    "tag_model": chat_tutor.model.tag,
                    "resposta": resposta,
                    "feedback_tutor": feedback_tutor
                })
                
                #print(feedbacks_questions)


            resp = {
                'id_person' : person["id_person"],
                'tag_model_person': person["tag_model_person"],
                'tag_model_tutor': chat_tutor.model.tag,
                'questions': questions_serializable,
                'resp_person': person["resp"],
                'feedback': feedbacks_questions
            }

            #print(resp)
            # Criar um campo "resp" com os valores retornados por state["answer"].
            resp_tutor.append(resp)

    # Crie um arquivo .json que armazene os dados da variável item.
    # Salvar o dataset de participantes em um arquivo JSON

    # Criar diretório se não existir
    dataset_dir = './'
    os.makedirs(dataset_dir, exist_ok=True)

    # Salvar o dataset em JSON
    output_path = os.path.join(dataset_dir, f'tutor_feedbacks.json')
    with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(resp_tutor, f, ensure_ascii=False, indent=2)
```
